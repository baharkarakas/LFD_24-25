```python
from pathlib import Path
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder

# 1. Veri setini yükleme
base_dir = Path.cwd().parent.parent
file_path = base_dir / 'LearningFromDataPrj' /'notebooks'/ 'merged_dataset_inorder.csv'
df = pd.read_csv(file_path)

# 2. `inf` ve `nan` değerlerini düzenleme
# `inf` değerlerini sütunun maksimum/minimum değeriyle değiştirin
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# `nan` değerlerini doldurma
for column in df.columns:
    if df[column].dtype in ['float64', 'int64']:  # Sayısal sütunlar için
        df[column] = df[column].fillna(df[column].median())  # Ortanca ile doldurma
    elif df[column].dtype == 'object':  # Kategorik sütunlar için
        df[column] = df[column].fillna(df[column].mode()[0])  # En sık görülen değer ile doldurma

# 3. Kategorik sütunları encode etme
categorical_columns = df.select_dtypes(include=['object']).columns

# Label Encoding
label_encoders = {}
for col in categorical_columns:
    label_encoders[col] = LabelEncoder()
    df[col] = label_encoders[col].fit_transform(df[col])

# 4. Özellik mühendisliği
# Mevcut özelliklerden türetilmiş yeni özellikler
for column in df.select_dtypes(include=['float64', 'int64']).columns:
    df[f'{column}_squared'] = df[column] ** 2  # Örneğin, sütunların karesi

# Standartlaştırma
scaler = StandardScaler()
scaled_columns = df.select_dtypes(include=['float64', 'int64']).columns
df[scaled_columns] = scaler.fit_transform(df[scaled_columns])

# 5. Sonuçları kaydetme
output_path = 'processed_merged_dataset_indorder.csv'
df.to_csv(output_path, index=False)

print(f"Veri seti başarıyla işlendi ve {output_path} dosyasına kaydedildi.")
```
